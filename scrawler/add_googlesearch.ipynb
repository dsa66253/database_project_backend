```python
abbr = gf['abb'].tolist()
print(abbr)
goo = gf['google_search'].tolist()
loc = df['t_office'].tolist()
google=[]
for i in range(len(loc)):
    k=0
    for j in range(len(abbr)):
        if(loc[i][0:2]==abbr[j]):
            google.append(goo[j])
            k=2
    if k<2:
        for j in range(len(abbr)):
            
            if(loc[i][0:1]==abbr[j]):
                print(loc[i],abbr[j])
                google.append(goo[j])
                k=1
                break;
    if(k<1):
        google.append('null')
```

```python
df[['google_search']] = pd.DataFrame(google)
```

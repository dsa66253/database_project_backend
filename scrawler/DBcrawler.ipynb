```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
```

```python
# 開啟瀏覽器視窗(Chrome)
# 方法一：執行前需開啟chromedriver.exe且與執行檔在同一個工作目錄
driver = webdriver.Chrome()
```

```python
driver.get("https://nol.ntu.edu.tw/nol/coursesearch/search_for_02_dpt.php?current_sem=111-1&coursename=&dptname=9010&dpt_sel=9000&yearcode=0&teachername=&selcode=-1&alltime=yes&week1=&week2=&week3=&week4=&week5=&week6=&allproced=yes&proced0=&proced1=&proced2=&proced3=&proced4=&procedE=&proced5=&proced6=&proced7=&proced8=&proced9=&procedA=&procedB=&procedC=&procedD=&allsel=yes&selCode1=&selCode2=&selCode3=&page_cnt=15")
```

```python
CID = []
cid_en = []
c_name = []
c_en_name = []
t_name = []
t_en_name = []
time = []
en_time = []
design_for = []
design_en_for = []

for i in range (16):
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    titles = soup.find_all('tr', {'align': 'center'})
    for title in titles:

        post = title.find_all('td')
        if post[0].getText().isdigit() and post[4].getText()!='\xa0' and post[10].getText() and (post[12].getText()!='\xa0'):
            #print(post[4].getText())
            CID.append(post[0].getText())
            design_for.append(post[1].getText())
            c_name.append(post[4].getText())
            t_name.append(post[10].getText())
            time.append(post[12].getText())

    driver.find_element(By.LINK_TEXT, "下一頁>>").click()
        
 
```
